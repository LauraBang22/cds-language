### Task 1 - Print all strings in capitalletters and two !

```python
#One solution
for colour in colours:
    print(colour.upper()+"!!")
```

```python
#Second solution
for colour in colours:
    shouted_colour = colour.upper()+"!!"
    print(shouted_colour)

#Giver mere transperens i hvordan man har gjort
```

```python
shouted_colours = []

#for each colour in the list "colours"
for colour in colours:
    shouted_colour = colour.upper()+"!!" #make upper case and add "!!" 
    shouted_colours.append(shouted_colour) #add this to the list
    print(shouted_colour) #print to screen
```

```python
#Tjek at koden over virker
print(shouted_colours)
```

### if/else

```python
#En type af statement
#Finder på conditions som skal opnåes, og hvis de er opfyldt sker der en ting, og hvis de ikker er opfyldt, så sker der noget andet

for colour in colours:
    if colour == "green":
        print(colour.upper()+"!!") #Har indtil videre kun sagt hvad der skal ske hvis det er ordet "green"
    else:
        print(colour) #hvis det ikke er "green", så skal der ikke gøres noget ved dem

```

```python
#Add seperate set of condition

for colour in colours:
    if colour == "green":
        print(colour.upper()+"!!")
    elif colour == "red":
        print("*"+colour+"*")
    else:
        print(colour)
```

```python
#En ny liste

weird_list = ("red", "green", 10, "blue")
```

```python
for value in weird_list:
    print(value/2) #får en error fordi vi ikke kan dividere alle values i listen
```

```python
#exception handling

for value in weird_list:
    if type(value) == int:
        print(value/2)
    else:
        print("Not a number!")

#Så får man ikke en error som oven over
#en anden mulighed er også bare at skrive "pass" under "else:"
```

### Exception handling

```python
#Når man vil forsøge og se hvad der sker
for value in weird_list:
    try:
        print(value/2)
    except:
        print("Not a value!")

#Lidt mere simpelt end kodeblokken over
#Man sætter ikke nogle faste konditions, men bare skriver hvad man gerne vil prøve
```

### Document length

```python
#split the string by spaces

tokens = text.split() #Hvis man ikke specificere hvor man skal spillet så splitter den på "white spaces"
```

```python
print(tokens)
```

### Counting word frequencies

```python
#count how many times a specif word appears
#define a word we are interested in

keyword = "love"
```

```python
tokens.count(keyword)
```

```python
#make lowercase 
tokens = text.lower().split()
```

```python
tokens.count(keyword)
```

```python
#remove punctuations
#import a new package

import string
```

```python
string.punctuation
```

```python
#list comprehension
cleaned_tokens = [''.join(char for char in token if char not in string.punctuation) for token in tokens]
```

```python
cleaned_tokens.count(keyword)
```
